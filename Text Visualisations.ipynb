```python
from nltk.corpus import brown
from nltk.corpus import stopwords
topics = ['government', 'news', 'religion', 'adventure', 'hobbies']
stop_words=stopwords.words("english")
for i in topics:
    text = [word for word in brown.words(categories =i) if word.lower() not in stop_words and word.isalpha()]
    freqdist = nltk.FreqDist(text)
    plt.figure(figsize = (16,12))
    freqdist.plot(50)
```

```python
from nltk.corpus import brown
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
import nltk
import numpy as np

def lexical_diversity(text):
    return len(set(text))/ len(text)

def get_brown_Id(category):
    word = [ word for word in brown.words(categories=category) if word.isalpha()]
    lex = lexical_diversity(word)
    return(category, lex)

for i in brown.categories()[:5]:
    word = [ word for word in brown.words(categories=i) if word.isalpha()]
    length =len(word)
    wordoffset = np.arange(2000, int(length/2000)*2000,2000)
    ld =[]
    for j in wordoffset:
        chunk = word[j-2000:j-1]
        ld.append(lexical_diversity(chunk))
    plt.plot(wordoffset, ld, label =i)
plt.legend()
plt.xlabel('word')
plt.ylabel('lexical Divesrsity')
plt.show()
```
